# Pymaceuticals Inc

## Observable Trends

* The tumor volume decreased over time for Capomulin, but increase for Infubinol, Ketapril, and Placebo at similar rates.
* Metestatic spreads for all the drugs were somewhat lower than the Placebo, with Ketapril, Infubinol, and Capomulin becoming progressively more effective and reducing the spreading.
* Only Capomulin distanced itself from the Placebo when comparing survival rates.
* When comparing the Tumor change in a 45 day period it becomes increasingly clear that only Capomulin decreased the Tumor size while Infubinol and Ketapril had similar changes to the placebo.

```python
# Dependencies and Setup
#%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from scipy import stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load
mouse_drug_data_to_load = os.path.join("data", "mouse_drug_data.csv")
clinical_trial_data_to_load = os.path.join ("data","clinicaltrial_data.csv")

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinic_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mice_clinic = pd.merge(mouse_data, clinic_data, on="Mouse ID", how="outer")


# Display the data table for preview
mice_clinic.head()
```

```python
mice_clinic.dtypes
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor = mice_clinic.groupby(["Drug", "Timepoint"]).mean()[["Tumor Volume (mm3)"]]

# Convert to DataFrame
mean_tumor = mean_tumor.reset_index (drop = False)

# Preview DataFrame
mean_tumor.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame

# Preview DataFrame

sem_tumor = mice_clinic.groupby(["Drug", "Timepoint"]).sem()[["Tumor Volume (mm3)"]]

# Convert to DataFrame
sem_tumor = sem_tumor.reset_index (drop = False)

# Preview DataFrame
sem_tumor.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_summary = mean_tumor.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
#mean_summary.keys.name = None
error_summary = sem_tumor.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")

# Preview that Reformatting worked
mean_summary.head()
#error_summary.head()
```

```python
# Create the data for the four lines and errors
Cap = mean_summary["Capomulin"]
Inf =mean_summary["Infubinol"]
Ket =mean_summary["Ketapril"]
Pla = mean_summary["Placebo"]

Cap_err = error_summary["Capomulin"]
Inf_err =error_summary["Infubinol"]
Ket_err =error_summary["Ketapril"]
Pla_err = error_summary["Placebo"]

x_axis = Cap.index
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(x_axis, Cap, yerr=Cap_err, color = "red", marker = "o", linestyle = ":", alpha = 0.5)
plt.errorbar(x_axis, Inf, yerr=Inf_err, color = "blue", marker = "^", linestyle = ":", alpha = 0.5)
plt.errorbar(x_axis, Ket, yerr=Ket_err, color = "green", marker = "s", linestyle = ":", alpha = 0.5)
plt.errorbar(x_axis, Pla, yerr=Pla_err, color = "black", marker = "d", linestyle = ":", alpha = 0.5)

plt.xlabel ("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title ("Total Response to Treatment")
plt.grid()
plt.legend()

# Save the Figure
plt.savefig("Tumor_Response.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_mat = mice_clinic.groupby(["Drug", "Timepoint"]).mean()[["Metastatic Sites"]]

# Convert to DataFrame
mean_mat = mean_mat.reset_index (drop = False)

# Preview DataFrame
mean_mat.head()

```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_mat = mice_clinic.groupby(["Drug", "Timepoint"]).sem()[["Metastatic Sites"]]

# Convert to DataFrame
sem_mat = sem_mat.reset_index (drop = False)

# Preview DataFrame
sem_mat.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_mat_summary = mean_mat.pivot(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")
#mean_summary.keys.name = None
error_mat_summary = sem_mat.pivot(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")

# Preview that Reformatting worked
mean_mat_summary.head()
#error_summary.head()
```

```python
# Create the data for the four lines and errors
Cap2 = mean_mat_summary["Capomulin"]
Inf2 =mean_mat_summary["Infubinol"]
Ket2 =mean_mat_summary["Ketapril"]
Pla2 = mean_mat_summary["Placebo"]

Cap_err2 = error_mat_summary["Capomulin"]
Inf_err2 =error_mat_summary["Infubinol"]
Ket_err2 =error_mat_summary["Ketapril"]
Pla_err2 = error_mat_summary["Placebo"]

x_axis2 = Cap2.index
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(x_axis2, Cap2, yerr=Cap_err2, color = "red", marker = "o", linestyle = ":", alpha = 0.5)
plt.errorbar(x_axis2, Inf2, yerr=Inf_err2, color = "blue", marker = "^", linestyle = ":", alpha = 0.5)
plt.errorbar(x_axis2, Ket2, yerr=Ket_err2, color = "green", marker = "s", linestyle = ":", alpha = 0.5)
plt.errorbar(x_axis2, Pla2, yerr=Pla_err2, color = "black", marker = "d", linestyle = ":", alpha = 0.5)

plt.xlabel ("Treatment Duration (Days)")
plt.ylabel("Met Sites")
plt.title ("Metastatic Spread During Treatment")
plt.grid()
plt.legend()

# Save the Figure
plt.savefig("Met.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_cnt = mice_clinic.groupby(["Drug", "Timepoint"]).nunique()[["Mouse ID"]]

# Convert to DataFrame
mice_cnt = mice_cnt.reset_index (drop = False)

# Preview DataFrame
mice_cnt.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mice_cnt_summary = mice_cnt.pivot(index = "Timepoint", columns = "Drug", values = "Mouse ID")

# Preview that Reformatting worked
mice_cnt_summary.head()
```

```python
# Create the data for the four lines and errors
Cap3 = mice_cnt_summary["Capomulin"]/mice_cnt_summary["Capomulin"].max() * 100
Inf3 =mice_cnt_summary["Infubinol"]/mice_cnt_summary["Infubinol"].max()* 100
Ket3 =mice_cnt_summary["Ketapril"]/mice_cnt_summary["Ketapril"].max()* 100
Pla3 = mice_cnt_summary["Placebo"]/mice_cnt_summary["Placebo"].max()* 100

x_axis3 = Cap3.index
```

```python
# Generate the Plot (Accounting for percentages)


# Save the Figure

# Show the Figure
plt.show()

plt.plot(x_axis3, Cap3, color = "red", marker = "o", linestyle = ":", alpha = 0.75)
plt.plot(x_axis3, Inf3, color = "blue", marker = "^", linestyle = ":", alpha = 0.75)
plt.plot(x_axis3, Ket3, color = "green", marker = "s", linestyle = ":", alpha = 0.75)
plt.plot(x_axis3, Pla3, color = "black", marker = "d", linestyle = ":", alpha = 0.75)

plt.xlabel ("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.title ("Survival During Treatment")
plt.grid()
plt.legend()

# Save the Figure
plt.savefig("Surv.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
drug_tumor = mean_tumor.set_index ("Drug")
Tumor_change =  ((drug_tumor.loc[drug_tumor["Timepoint"] == 45, "Tumor Volume (mm3)"]/drug_tumor.loc[drug_tumor["Timepoint"] == 0, "Tumor Volume (mm3)"] )-1)*100

# Display the data to confirm
Tumor_change
```

```python
# Store all Relevant Percent Changes into a Tuple
summary_df = pd.DataFrame({"Change": Tumor_change.values}, index = Tumor_change.index)

drug_names = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
Subset = summary_df.loc[drug_names, :]
Tumor_tupl = tuple(Subset.values.T.tolist())


# Splice the data between passing and failing drugs
bins = [-100, 0, 100]
group_names = ["Passing", "Failing"]
Subset["Pass/Fail"] = pd.cut(Subset["Change"], bins, labels = group_names) 
data = Subset["Change"]
data_tupl = tuple(data)

# # Orient widths. Add labels, tick marks, etc. 
fig, ax = plt.subplots()
width = 1 
x_axis4 = np.arange(len(data))
colors = np.where(data >= 0, 'red', 'green')

ax.bar(x_axis4, data, width, color=colors, align="edge")
tick_locations = [value + 1 for value in x_axis4]
ax.set_xlim(-0.25, 4.25)
ax.set_ylim(Subset["Change"].min()-5, Subset["Change"].max()+10)
plt.xticks(tick_locations, drug_names)
plt.title("Tumor Change over 45 Day Treatment")
plt.xlabel("Drugs")
plt.ylabel("% Tumor Volume Change")
plt.grid()


# Use functions to label the percentages of changes
def pct_label(tupl):
    for pct in tupl:
        plt.text(data_tupl.index(pct)+.33, pct / 2, "{0:.0%}".format(pct/100), color='white', va = "center")

# Call functions to implement the function calls
pct_label (data_tupl)

# Save the Figure
plt.savefig("Tumor_Change.png")

# Show the Figure
plt.show()
```
